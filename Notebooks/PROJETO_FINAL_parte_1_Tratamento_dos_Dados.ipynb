### **2.1 FUN√á√ïES**

```python
'''
Essa Fun√ß√£o retorna a quantidade de dados vazios de um dataframe
'''
def valores_nulos (dados):
  '''
  Par√¢metros
  ----------
  dados: dataframe
  '''
  #Quatindade de Valores Nulos  
  valores_nulos = dados.isna().sum().sum()
  
  #Mostrando a Quantidade de Valores Nulos
  print(f'Esse dataset possui {valores_nulos} valores nulos')
```

```python
'''
Essa Fun√ß√£o identifica o grupo de features continuas, categ√≥ricas e os dados de sa√≠da. Al√©m disso preenhce os dados nulos 
'''
def preenche_tabela(dados):
  '''
  Par√¢metros
  ----------
  dados: dataframe
  '''
  #Separar as features continuas
  features_continuas_colunas = dados.iloc[:, 13:-2].columns
  
  #preenchendo os valores vazios
  features_continuas = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index=False)[features_continuas_colunas].fillna(method='bfill').fillna(method='ffill')
  features_categoricas = dados.iloc[:, :13]
  
  #Separar as features de sa√≠da
  saida = dados.iloc[:,-2:] #selecionando as linhas
  dados_finais = pd.concat([features_categoricas, features_continuas, saida], ignore_index = True, axis=1) #definir eixo=1
  dados_finais.columns = dados.columns
  return dados_finais
```

```python
'''
Fun√ß√£o que define as duas primeiras horas como janela de tempo e retira os pacientes que ja chegaram direto na UTI
'''
def prepare_window(rows):
  '''
  Par√¢metros
  ----------
  rows: linhas que s√£o respectivas a um paciente
  '''
  if(np.any(rows['ICU'])):
    rows.loc[rows['WINDOW']=='0-2', 'ICU'] = 1
  return rows.loc[rows['WINDOW']=='0-2']
```

```python
'''
Fun√ß√£o que define as anota√ß√µes em um plot, essa fun√ß√£o foi desenvolvida pelo github(https://github.com/Glaudemias/storytelling-with-data)
'''
def anotacao(x, y, text, colors, spacing=20, ax=None, **kw):
    colors = list(reversed(colors))
    t = ax.transData
    canvas = ax.figure.canvas

    for i, line in enumerate(reversed(text.split('\n'))):
        strings = line.split('||')
        for s, c in zip(strings, colors[i]):
            text = ax.text(x, y, s, color=c, transform=t, **kw)
            text.draw(canvas.get_renderer())
            ex = text.get_window_extent()
            t = transforms.offset_copy(text._transform, x=ex.width, 
                                       units='dots')

        t = transforms.offset_copy(ax.transData, x=0, y=(i + 1) * spacing, 
                                   units='dots')
```

```python
'''
Fun√ß√£o que remove as features de maior correla√ß√£o de um data set
'''
def remove_corr_var(dados,alta_corr):
  '''
  Par√¢metros
  ----------
  dados: dataframe
  alta_corr = Valor de alta correla√ß√£o
  '''

  #Criando a Matriz de Correla√ß√£o
  matrix_corr = dados.iloc[:,4:-2].corr().abs() 
  #Criando a Matriz de Correla√ß√£o, eliminando um dos triangulos e a diagonal principal
  matrix_TRIsuperior = matrix_corr.where(np.triu(np.ones(matrix_corr.shape),k = 1).astype(np.bool))
  excluir = [coluna for coluna in matrix_TRIsuperior.columns if any(matrix_TRIsuperior[coluna]>alta_corr)]

  return dados.drop(excluir,axis=1)
```

Com a base de dados ok, vamos definir os seguintes crit√©rios:

* A alta correla√ß√£o a partir de <font color='darkorange'>**0.95**,<font color='black'>
* Utilizando o m√©todo `np.bool` para trocar os valores para booleanos `np.bool`
* O m√©todo `where`ir√° trocar os valores onde a condi√ß√£o for False e eliminar onde a condi√ß√£o `np.ones` for falsa
* Iremos tentar eliminar a diagonal principal e tri√¢ngulo inferior da nossa matriz usando os m√©todos `np.triu` e k=1
* Iremos plotar e visualizar melhor esse triangulo atrav√©s de uma plotagem dessa matriz de correla√ß√£o.

**O total de colunas a serem eliminadas**

```python
#Observando a quantidade de colunas a ser excluida 

alta_corr = 0.95

#Criando a Matriz de Correla√ß√£o
matrix_corr = dados_limpos.iloc[:,4:-2].corr().abs() 
#Criando a Matriz de Correla√ß√£o, eliminando um dos triangulos e a diagonal principal
matrix_TRIsuperior = matrix_corr.where(np.triu(np.ones(matrix_corr.shape),k = 1).astype(np.bool))
colunas_excluir = [coluna for coluna in matrix_TRIsuperior.columns if any(matrix_TRIsuperior[coluna]>alta_corr)]
print(f'o Total de colunas a ser excluido √© {len(colunas_excluir)}')
```

**Visualizando a Correla√ß√£o das informa√ß√µes**

```python
matrix_corr = dados_limpos.iloc[:,4:-2].corr().abs() 

#Criando a Matriz de Correla√ß√£o, eliminando um dos triangulos e a diagonal principal
matrix_TRIsuperior = np.triu(np.ones(matrix_corr.shape),k = 1).astype(np.bool)

#Total de colunas a ser excluida

fig,ax = plt.subplots(1,figsize=(20,12))
ax=sns.heatmap(data=matrix_corr,cmap='RdPu', mask= matrix_TRIsuperior, fmt='.2f')
plt.title('$\\bf{CORRELA√á√ÉO \ DOS\ DADOS\ }$\nVisualizando a Correla√ß√£o de Pearson', fontsize= 24, loc= 'left', color = 'gray')

#mostrando o total de colunas a ser excluidas
anotacao(110.6,50,
             'De acordo com a ||$\\bf{correla√ß√£o}$\n'
             '$\\bf{das\ features}$|| o total de colunas\n'
             'a serem eliminadas s√£o: ||$\\bf{131}$',
             [['dimgray', 'rebeccapurple'],
              ['rebeccapurple','dimgray'],
              ['dimgray','rebeccapurple']], 
             spacing=20,
             ax=ax,
             fontsize=24)
```
